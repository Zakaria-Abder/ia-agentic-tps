```python
from dotenv import load_dotenv
from tavily import TavilyClient
from langchain_core.tools import tool
from langchain_core.messages import SystemMessage, HumanMessage
from langgraph.checkpoint.memory import InMemorySaver
from langgraph.prebuilt import create_react_agent
from langchain_ollama import ChatOllama
```

```python
import tavily
print(tavily.__file__)
```

```python
load_dotenv()
```

```python
model = ChatOllama(model="llama3.2:3b")
```

```python
tavily_client = TavilyClient()
```

```python
@tool
def web_search(query: str):
    """
    Recherche des informations culinaires sur Internet.
    """
    return tavily_client.search(query=query)["results"]
```

```python
system_message = SystemMessage(content="""
Tu es un chef cuisinier personnel intelligent.

Tes tâches sont :
- mémoriser les préférences de l'utilisateur ;
- recevoir les ingrédients disponibles ;
- rechercher sur le web si nécessaire ;
- proposer des recettes adaptées ;
- expliquer brièvement les étapes de préparation ;
- éviter les ingrédients que l'utilisateur n'aime pas.
""")
```

```python
agent = create_react_agent(
    model=model,
    tools=[web_search],
    checkpointer=InMemorySaver()
)
```

```python
config = {"configurable": {"thread_id": "1"}}
```

```python
question = HumanMessage(
    content="""
Bonjour.
Je n'aime pas les champignons.
Je préfère les plats épicés.
"""
)

agent.invoke(
    {"messages": [system_message, question]},
    config
)
```

```python
question = HumanMessage(
    content="""
Dans mon réfrigérateur j'ai :

- poulet
- tomates
- oignons
- pommes de terre
- ail

Propose-moi un plat.
"""
)

response = agent.invoke(
    {"messages": [question]},
    config
)

print(response["messages"][-1].content)
```

```python
question = HumanMessage(
    content="""
Je veux une recette marocaine à base de poulet et pommes de terre.
"""
)

response = agent.invoke(
    {"messages": [question]},
    config
)

print(response["messages"][-1].content)
```
